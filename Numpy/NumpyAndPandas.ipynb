```python
import numpy as np
import pandas as pd
```

```python
honeyproduction_US = pd.read_csv('honeyproduction.csv')
```

```python
honeyproduction_US.head()
```

```python
total_prod = np.sum(honeyproduction_US['totalprod'])
total_prod
```

```python
honeyproduction_US = honeyproduction_US.dropna(how = 'any')
honeyproduction_US.describe()
```

```python
honeyproduction_US["state"].value_counts()
```

```python
honeyproduction_US = honeyproduction_US.drop(columns = ['unnamed'])
```

```python
year_wise = honeyproduction_US.groupby('year').aggregate(np.mean)
```

```python
year_wise
```

```python
year_wise = year_wise.drop(columns=['yieldpercol','priceperlb','prodvalue'])
```
