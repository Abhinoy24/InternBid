```python
import pandas as pd
```

```python
students = pd.DataFrame({'id_number' : [100,120,101,131,221],
                         'Name' : ["Abhinoy Pradhan","Kiwi Mwchahary","Aroj Subedhi","Bishal Gurung","Aweshis Katel"],
                         'Age' : [22,20,29,25,30],
                         'Gender' : ['M','F','M','M','M']})
```

```python
students
```

```python
professors = pd.DataFrame({'Prof_Name':["Sandeep Gurung","Nitisha Pradhan","Reena Pradhan","Dheeraj Kumar","Leeon Pandi"],
                           'id_number' : [100,101,120,131,212],
                           'Department' : ["CSE","CSE","Civil","ME","ME"]})
```

```python
professors
```

```python
pd.concat([students,professors],sort=True)
```

```python
pd.concat([students,professors],ignore_index=True,sort=True)
```

```python
pd.merge(students,professors)
```

```python
pd.merge(students,professors, on = 'id_number')
```

```python
pd.merge(students,professors, on = 'id_number', how = 'left')
```

```python
 pd.merge(students,professors, on = 'id_number' , how = 'right')
```

```python
pd.merge(students,professors, on = 'id_number' , how = 'inner')
```

```python
pd.merge(students,professors, on = 'id_number' ,how ='outer')
```
